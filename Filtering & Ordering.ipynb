# filter()
* In pandas, the filter() function is used to subset or filter a DataFrame based on the labels (columns or index levels) along a specified axis. It allows you to select a subset of columns or rows based on specific criteria.
# DataFrame.filter(items=None, like=None, regex=None, axis=None)
# Parameters-* items: A list of column or index names to include. If None, all items that match the criteria will be included.
* 
like: A string to match against the column or index names using a partial strin match.
* regex: A regular expression to match against the column or index names.
* axis: The axis to filter items along (0 for rows/index, 1 for columns).

```python
# print capitals of the countries
df2.filter(items=["Capital"], axis=1)
```

```python
# print India-China info
df2.filter(items=["India", "China"], axis=0)
```

```python
df2.loc[["India", "China"]]
```

```python
# filter Vs loc
# df2.loc[["India", "Capital"]] error
# Both are used to filter the dataframes but filter() works on index & column whereas loc
# works only on index.
```

# Sorting
1. sort_values()
2. sort_values(by='column')
3. sort_values(by='column', ascending=False) --> sorts in descending order
4. sort_values(by=['column1', 'column2'], ascending=False) --> sort according to column1 & column2 . Priority decreases while going from left to right and it will be sorted in descending order. 
5. sort_values(by=['column1', 'columns2'], ascending=[False, True]) --> sort column1, in descending whereas column2 in ascending.

```python
df[df["Rank"] <= 10].sort_values(by='Rank')
```

```python
df[df['Rank'] <= 20].sort_values(by='Rank', ascending=False)
```

```python
df[df['Rank'] <= 10].sort_values(by=["Capital", "Continent"], ascending=False)
```

```python
df[df['Rank'] <= 10].sort_values(by=["Capital", "Continent"], ascending=[False, True])
```
